```python
# Dicionários
meudici = {
  "nome": "PL",
  "curso": "LEI",
  "ano": 2022
}
# Acesso
meudici["nome"]

# Chaves
for key in meudici.keys():
    print("key",key)

# Valores
for value in meudici.values():
    print("value",value)

# Iterar
for key, value in meudici.items():
    print("key",key,value)
```
